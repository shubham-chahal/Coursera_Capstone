```python
# Defining the FourSquare API
search_query = 'Mall'
LIMIT = 50 # Used 50 limit, since all unique ids obtained would be called and my free account only allows 50 premium calls per day
url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, search_query, LIMIT)
url
```

```python
# Sending the GET Request and examining the results
results = requests.get(url).json()
```

```python
# Assigning relevant part of JSON to venues
venues = results['response']['venues']

# Tranforming venues into a dataframe
df = json_normalize(venues)
df.head()
```

```python
id_list=df["id"] # These 50 ids would be called by Foursquare API
name_list=df['name']
rating_list=[]
for i in range(50): # Loop to call all malls ids
    venue_id=id_list[i]
    url = 'https://api.foursquare.com/v2/venues/{}?client_id={}&client_secret={}&v={}'.format(venue_id, CLIENT_ID, CLIENT_SECRET, VERSION)
    result = requests.get(url).json()
    try:
        rating=result['response']['venue']['rating'] #storing rating result, if present
    except:
        rating=0 # Marking the rating as 0 if not available
    rating_list.append(rating)   # Making a rating list
```

```python
print(name_list)
print(rating_list)
```

```python
Rating_Dict=dict(zip(name_list, rating_list)) # Making a dictionary, with index as mall name and rating as value
print(Rating_Dict)
```

```python
Highest_Rating=0
for k in Rating_Dict: # Loop to find the highest rated mall
    if Rating_Dict[k]>Highest_Rating:
        Highest_Rating=Rating_Dict[k]
        Highest_Rating_Mall=k
    else:
        pass
print(Highest_Rating_Mall,Highest_Rating)   
```

```python
Highest_Rating_Malls=[]
for k in Rating_Dict: # Loop to check if multiple malls are present with above calculated highest rating
    if Rating_Dict[k]==Highest_Rating:
        Highest_Rating_Malls.append(k)
    else:
        pass
print(Highest_Rating_Malls) # no dublicacy in highest rating
```
